```python
#OPRATION ON LIST
l1 = list([2,3,1,4,5])
l1.sort()
print(l1)

l1.append(0.1)
print(l1)
l1.insert(2,32)
print(l1)
l1.sort()
l1.pop()
print(l1)
l1.reverse()
print(l1)
l1.index(0.1)
```

```python
#Dictonary dataType
d1 ={"name":"Chaitanya","age":19,"college":"GPCS"}
print(d1)
print(type(d1))

#Using set as a key in Dictonary
s1  = {1,2,3}
print(type(s1))
s1 = frozenset(s1)
print(type(s1))
print(s1)
#dict2 = {s1 :[2,3,45]}
dict2 = dict(s1=[2,3,45])
print(dict2)
#Accessing the vales , keys and items of the dictonary data type
print(type(dict2))
print(dict2.values())
print(dict2.keys())
print(dict2.items())
```
